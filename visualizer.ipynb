```python
import numpy as np
import plotly.graph_objects as go

# Using readlines() 
file1 = open('day2night/other_data/loss_log.txt', 'r') 
Lines = file1.readlines() 
  
count = 1
index2 = []

#loss data
ggan = []
g_l1 = []
d_real = []
d_fake = []

epoch = []

time = 0

for line in Lines: 
    res = line.split()   
    ep = res[1]
    ep = ep[:-1]
    
    t = res[1]
    t = t[:-1]
    time = time + float(t) 
    epoch.append(int(ep))
    
    #read data
    ggan.append(float(res[9]))
    g_l1.append(float(res[11]))
    d_real.append(float(res[13]))
    d_fake.append(float(res[15]))
    
    index2.append(count)
    count = count + 1


control = epoch[0]  
epochs = []  
epoch_count = 0
for i in range(len(epoch)):
    if epoch[i] != control:
        epoch_count = epoch_count + 1
        epochs.append(epoch_count)
        control = epoch[i]
        
epochs = np.asarray(epochs, dtype=int)
index2 = np.asarray(index2, dtype=float)

index2 = index2 / 175


# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=index2, y=g_l1,
                    mode='lines',
                    name='GL1'))
fig.add_trace(go.Scatter(x=index2, y=ggan,
                    mode='lines',
                    name='GGAN'))
fig.add_trace(go.Scatter(x=index2, y=d_real,
                    mode='lines',
                    name='D_Real'))
fig.add_trace(go.Scatter(x=index2, y=d_fake,
                    mode='lines',
                    name='D_Fake'))

fig.show()

hours = time/3600
seconds = time % 3600
#print("Hours: "+ str(hours) + " seconds: " + str(seconds))
```
